```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline # allows you to show the plots in your Jupyter notebook
# plt.show()
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional
plt.plot(x, y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# if you're not in jupyter notebook you'll have to say plt.show() though it can also work in Jupyter notebook
```

```python
plt.plot(x, y, 'r--')
```

```python
plt.subplot(1, 2, 1) # order is number of rows, columns and the plot number you're referring to
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # order - left, bottom, width and height

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # the numbers have to be between zero and one
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # 20% from the left, 50% from the bottom, width of 40% and height of 30%

axes1.plot(x, y)
axes1.set_title('LARGER PLOT')

axes2.plot(y, x)
axes2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

#axes.plot(x, y)
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3)

plt.tight_layout() # fixes the overlaps of the plots because normally they would be squished
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2) # axes is an array of axes

for current_ax in axes:
    current_ax.plot(x, y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2) 

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(x,y)
axes[1].set_title('Second Plot')
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(3, 2)) # the tuple contains the width and height of the figure in inches
# fig = plt.figure(figsize=(3, 2), dpi=100) second argument which is dots per inch or pixels per inch

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(8, 2)) 

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(figsize=(8, 2))

axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))

axes[0].plot(x, y)

axes[1].plot(y, x)

plt.tight_layout()
```

```python
# saving a figure to a file
fig.savefig('my_picture.png') # or fig.savefig('my_picture.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend() # ax.legend(loc=0) specifies the location where your legend should be, loc=0 specifies that matplotlib should place it in the best possible location
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc=(0.1,0.1))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc=10) # center - just check documentation guide
```
