```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris) # automated jointplot for the entire dataset
```

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot) # there's a dist plot on the diagonal of the plots, not dist plot is deprecated, use histplot instead 
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot) # there's a dist plot on the diagonal of the plots, not dist plot is deprecated, use histplot instead 
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
