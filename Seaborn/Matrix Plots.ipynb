```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
# In order for heatmaps to work properly, your data should already be in a matrix form
# meaning the index name and column name match up such that the cell value indicates something relevant for both of them
tips.corr()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True)
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')   
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=3)
```

```python
sns.clustermap(fp, cmap='coolwarm')
```
