```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips) # enables you to aggregate the categorical data based off an aggregate function
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips) # just counts occurrences
```

```python
# boxplot and violin plots shows distribution of categorical data
# for both x is category and y is a numerical column
# boxplot shows distribution of quantitative data that facilitates comparisons between variables
sns.boxplot(x='day', y='total_bill', data=tips) # boxplots - shows quartiles of the dataset and whiskers extend to show the distribution, points outside are outliers 
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips) # a boxplot is more easier to read than violin plots, violin plots give a lot more information
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips) # basically a scatterplot based off the category
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True) # helps in case you have a number of points stacked ontop of each other by adding a little noise
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True) # the split parameter was named to dodge
```

```python
# swarm plot is combination of stripplot and violin plot
# points are adjusted such that they don't overlap
# NOTE: They usually don't always scale very well for very large datasets
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
# Not very useful for explanation but more useful for explaratory data analysis
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar') # kind describes the factor plot itself
```

```python
# The factorplot function changed to catplot
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```
